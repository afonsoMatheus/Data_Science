```python
import pandas as pd
import numpy as np
import heapq as hp
from scipy.spatial import distance
from statistics import mode
import seaborn as sns

data = pd.read_csv("iris.csv")
data.head()
```

```python
def knn(train, test, k):
        
    train = train.values.tolist()
    heap = []
    
    for i in range(len(train)):
        
        hp.heappush(heap, [distance.euclidean(train[i][:4], test), train[i][4]])
        
    near = heap[0:k]
    result = []
    
    for i in range(k):
        
        result.append(near[i][1])
        
    result = mode(result)
    
    return result
    
```

```python
sns.catplot(x="sepal.length", y="sepal.width", hue = "variety",
            kind="swarm", data=data, size = 10);

sns.catplot(x="petal.length", y="petal.width", hue = "variety",
            kind="swarm", data=data, size = 10);
```
